```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from scipy.stats import f_oneway
warnings.filterwarnings('ignore')
```

<h1>INSIGHT 1</h1>
<h2>WHICH YEAR RECORDED THE HIGHEST NUMBER OF DENGUE CASES?</h2>
<h3>THE YEAR RECORDED THE HIGHEST NUMBER OF DENGUE CASES IS 2019</h3>

```python
lamok.groupby('Year')['Dengue_Cases'].sum()

yearly_cases = lamok.groupby('Year')['Dengue_Cases'].sum()
plt.figure(figsize=(10,5))
plt.plot(yearly_cases.index, yearly_cases.values, marker='o', linestyle='-', color='blue', label="Dengue")
plt.xlabel('Year')
plt.ylabel('Total Dengue Cases')
plt.title('Total Dengue Cases Per Year')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```
