```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from scipy.stats import f_oneway
warnings.filterwarnings('ignore')
```

```python
internet = pd.read_csv("Internet-Demographic.csv")
```

<h1>1. DISTRIBUTION OF BIRTH RATES</h1>
<h2></h2>

```python
plt.figure(figsize=(10,5))
sns.histplot(internet["Birth rate"], bins=20)
plt.title("Distribution of Birth Rates")
plt.xlabel("Birth Rate per 1,000 people")
plt.ylabel("Frequency")
plt.show()
```

<h1>2. TOP 5 COUNTRIES WITH THE HIGHEST INTERNET USAGE</h1>
<h2> THE TOP 5 COUNTRIES WITH THE HIGHEST INTERNET USAGE IS ICELAND, BERMUDA, NORWAY, SWEDEN, AND DENMARK</h2>

```python
top_5 = internet.nlargest(5, 'Internet users')[['Country Name', 'Internet users']]
print(top_5)
```

<h1>3. COUNT OF COUNTRIES BY INCOME GROUP</h1>

```python
plt.figure(figsize=(10, 8))
sns.countplot(y=internet['Income Group'], palette='viridis', order=internet['Income Group'].value_counts().index)
plt.title('Count of Countries by Income Group')
plt.xlabel('Number of Countries')
plt.ylabel('Income Group')
plt.show()
```

<h1>4. BOX PLOT OF INTERNET USAGE (OPTIONAL)</h1>

```python
plt.figure(figsize=(10, 8))
sns.boxplot(y=internet['Internet users'], color='aqua')
plt.title('Count of Countries by Income Group')
plt.xlabel('Number of Countries')
plt.ylabel('Income Group')
plt.show()
```

<h1>5. COUNTRIES WITH NO INTERNET ACCESS</h1>

```python
no_internet = internet[internet['Internet users'] == 0][['Country Name']]
print("no_internet")
```

<h1>6. AVERAGE OF BIRTH RATE</h1>
<h2>THE AVERAGE OF BIRTH RATE IS 21%</h2>

```python
avg_birth_rate = internet['Birth rate'].mean()
print("Average Birth Rate:", avg_birth_rate)
```

<h1>7. SCATTER PLOT: BIRTH RATE VS INTERNET USERS</h1>

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x=internet['Internet users'], y=internet['Birth rate'])
plt.title('Scatter Plot for Birth Rate')
plt.xlabel('Internet Users (%)')
plt.ylabel('Birth Rate')
plt.show()
```

<h1>8. COUNTRIES WITH BIRTH RATES ABOVE 40</h1>

```python
high_birth_rate = internet[internet['Birth rate'] > 40][['Country Name', 'Birth rate']]
print("Countries with birth rates above 40:\n", high_birth_rate)
```

<h1>9. PIE CHART OF INCOME GROUPS</h1>

```python
plt.figure(figsize=(10, 8))
internet['Income Group'].value_counts().plot.pie(autopct='%1.1f%%', color=['aqua', 'lightblue', 'lightgreen', 'yellow'])
plt.title('Proportion of Income Groups')
plt.ylabel('')
plt.show()
```
