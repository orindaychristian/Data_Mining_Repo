```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from scipy.stats import f_oneway
warnings.filterwarnings('ignore')
```

```python
accident = pd.read_csv("accident_data.csv")
```

<h1>INSIGHT 1</h1>
<h2>QUESTION: HOW DO ACCIDENTS VARY UNDER DIFFERENT LIGHT CONDITIONS?</h2>
<h3>ANSWER: THE MAJORITY OF ACCIDENTS OCCUR IN DAYLIGHT</h3>

```python
accident['Light_Conditions'].value_counts()
light_conditions_counts = accident['Light_Conditions'].value_counts()
plt.figure(figsize=(10, 5))
light_conditions_counts.plot(kind='bar', color='aqua', edgecolor='black')

plt.xlabel('Light Conditions')
plt.ylabel('Count')
plt.title('Distribution of Light Conditions in Accidents')
plt.xticks(rotation=45)
plt.show()
```

<h1>INSIGHT 2</h1>
<h2>QUESTION: CORRELATION BETWEEN NUMBER OF CASUALTIES AND NUMBER OF VEHICLES?</h2>
<h3>ANSWER: THERE IS A CORRELATION BETWEEN NUMBER OF CASUALTIES AND NUMBER OF VEHICLES</h3>

```python
number_casualties_vehicles = accident['Number_of_Casualties'].corr(accident['Number_of_Vehicles'])
number_casualties_vehicles

number_casualties = accident['Number_of_Casualties']
number_vehicles = accident['Number_of_Vehicles']
number_casualties_vehicles = number_casualties.corr(number_vehicles)
labels = ['Number of Casualties', 'Number of Vehicles']
values = [number_casualties_vehicles]
plt.figure(figsize=(6, 4))
plt.bar(labels, values, color=['lightblue', 'lightgreen'])
plt.xlabel('Variables')
plt.ylabel('Correlation')
plt.title('Correlation Between Number of Casualties and Number of Vehicles')
plt.show()
```
